```python
import seaborn as sns
sns.get_dataset_names()
```

```python
df=sns.load_dataset("tips")
```

```python
print(df.head())
df.dtypes
```

```python
#corrleation it has feature heatmap
#which helps to find correleation between each and  every feature
#this correleation is very much imp wen v do feature selection

df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
# join plot
#a join plot allow us to study the relationship between 2  numerical variables the central chart
#displays their correlation
#it is usually a scatterplot, a hexbin plot, a 2d histogram or a 2d density plot
```

## UNIVARIATE ANALYSIS

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
#draw a pdf
#draw a reg line that is a best fit line
#curve line pdf
```

```python
#pair plot
#a pair plot is also known as scatter plot ,in which 1 variable in the same data row is matched with another variable value
#wen v have more independent features
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
#dist plot used to create histograms
sns.distplot(df['tip'])
#kde is curve shiwn
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
#kde is not shown kernel density  estimation
```

```python
# categorical feature
#count plot 
sns.countplot('sex',data=df)
```

```python
sns.countplot('day',data=df)
```

```python
#bar plot
sns.barplot(x='total_bill',y='sex',data=df)
```

```python
#boxplot
#a box and whisker plot(sometimes called a box plot) is a graph that presents info in a 5no summary
sns.boxplot('smoker','total_bill',data=df)
```

```python
sns.boxplot(x="total_bill",y="day",hue="smoker",data=df)
```

```python
#violin plot 
#helps to see both the distribution of data in terms of kernel density estimation and the boxplot
sns.violinplot(x="total_bill",y="day",data=df,palette='rainbow')
```

```python
#strip plot
sns.stripplot(x=df['tip'])
```

```python
sns.stripplot(x='day',y='total_bill',data=df)
```

```python
#facet grid
import matplotlib.pyplot as plt
x=sns.FacetGrid(df,row='smoker',col='time')
x=x.map(plt.hist,'total_bill')
```
